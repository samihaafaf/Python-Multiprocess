READING FILES RECURSIVELY AND STORING IN A LIST

```python
import os

#getting the files recursively
def find_text_files(directory):
    text_files = []
    for root, _, files in os.walk(directory):
        for file in files:
            if file.endswith(".txt"):
                text_files.append(os.path.join(root, file))
    return text_files

directory_path = "/Users/samihaafafneha/Desktop/mishkat/WORK2/10K"  # Replace this with the path of your folder
files = []
if os.path.exists(directory_path) and os.path.isdir(directory_path):
    txt_files = find_text_files(directory_path)
    if txt_files:
        for txt_file in txt_files:
            #print(f"Found .txt file: {txt_file}")
            files.append(txt_file)
#files

```
